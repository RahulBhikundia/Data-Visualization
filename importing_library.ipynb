Use any 2 column attribute for visualizing their relationship

```python
d3 = d1[['sepal_length','sepal_width','petal_length','petal_width']]
sns.heatmap(d3.corr())
```

from the above heatmap, it can be deduced that, petal length and petal width are highly correlated. So, it would be best to visualize relationship between them

```python
plt.boxplot(d3)
```
